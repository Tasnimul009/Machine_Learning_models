```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 100, 
                       n_features = 1, 
                       n_targets = 1,
                       n_informative = 1, 
                       noise = 50,
                       random_state = 7)
plt.scatter(X, y, 
            color = "yellow",
            edgecolor = "k",
            s=200)
```

```python
epochs = 100
lr = 0.1
b = -23
m = 1
all_b = []
all_m = []
for i in range(epochs):
    slope_m = np.mean(-2*(y - m*X - b) * X)
    slope_b = np.sum(-2*(y - m*X - b) )
    b -= lr * slope_b
    m -= lr * slope_m
    all_b.append(b)
    all_m.append(m)
np.array(all_b).shape
final_m = all_m[-1]
final_b = all_b[-1]
plt.scatter(X, y, 
            color = "yellow",
            edgecolor = "k",
            s=200)
plt.ylim(-200, 200)
plt.xlim(-3, 3)
plt.plot(X, final_m*X + final_b , 
         color = "blue", linewidth = 10)
```
