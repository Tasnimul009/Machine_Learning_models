```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print("hello world")
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 100,
                       n_features = 1,
                       n_targets = 1,
                       n_informative = 1,
                       noise = 50,
                       random_state = 6)
plt.scatter(X, y, color = "yellow", edgecolor = "k", s = 200)
print(X.shape)
print(y.shape)
```

```python
m_arr = np.linspace(-150, 150, 10)
b_arr = np.linspace(-150, 150, 10)
mg, bg = np.meshgrid(m_arr, b_arr)
final = np.vstack((mg.ravel(), bg.ravel())).T
final
z_arr = []
for i in range(final.shape[0]):
    z_arr.append(
        np.sum((y - final[i][0] * X - final[i][1])**2)
    )
z = np.array(z_arr).reshape(mg.shape)
```

```python
import plotly.graph_objects as go 
fig = go.Figure(data = [go.Surface(
    x = mg, y = bg, z = z
)])
fig.show()
title = "here_we_go.html"
fig.write_html(title)
import webbrowser as wb 
wb.open(title)
```
