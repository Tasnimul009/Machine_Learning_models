```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class SGD:
    def __init__(self, lr = 0.001, epochs = 100):
        self.coef_ = None
        self.intercept_ = None
        self.lr = lr
        self.epochs = epochs
    def fit(self, X, y):
        self.coef_ = 1
        self.intercept_ = 0
        for i in range(self.epochs):
            for j in range(X.shape[0]):
                idx = np.random.randint(0, X.shape[0])
                coef_der = -2 * (y[idx] - self.coef_*X[idx] - self.intercept_) * X[idx]
                intercept_der = -2 * (y[idx] - self.coef_*X[idx] - self.intercept_)
                self.coef_ -= self.lr * coef_der
                self.intercept_ -= self.lr * intercept_der 
        print(self.coef_, self.intercept_)
    def predict(self, X):
        y_pred = X * self.coef_ + self.intercept_
        return y_pred
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 100,
                       n_features = 1,
                       n_targets = 1,
                       n_informative = 1,
                       noise = 30,
                       random_state = 12)
plt.scatter(X, y)
```

```python
reg = SGD()
reg.fit(X, y)
y_pred = reg.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred)
```
