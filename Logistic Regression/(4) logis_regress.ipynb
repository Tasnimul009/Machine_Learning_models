```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples = 100,
                           n_features = 2,
                           n_informative = 1,
                           n_redundant = 0,
                           n_classes = 2,
                           n_clusters_per_class = 1,
                           hypercube = False,
                           random_state = 41,
                           class_sep = 10)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2,
                                                    random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred_1 = model.predict(X_train)
y_pred_2 = model.predict(X_test)
from sklearn.metrics import accuracy_score, confusion_matrix
ac = accuracy_score(y_test, y_pred_2)
cm = confusion_matrix(y_test, y_pred_2)
print("Accuracy:", ac)
print("Confusion Matrix:\n", cm)
```

```python
m = -(model.coef_[0][0] / model.coef_[0][1])
b = -(model.intercept_ / model.coef_[0][1])
```

```python
X_i = np.linspace(X_test[:, 0].min(), X_test[:, 0].max(), 100)
y_i = m * X_i + b
plt.scatter(X_test[:, 0], X_test[:, 1], c = y_pred_2, cmap = 'viridis')
plt.figure()
plt.xlim(X_test[:, 0].min(), X_test[:, 0].max())
plt.ylim(X_test[:, 1].min(), X_test[:, 1].max())
plt.scatter(X_test[:, 0], X_test[:, 1], c = y_test, cmap = 'cool', marker = 'x')
plt.plot(X_i, y_i, color = 'red')
```
