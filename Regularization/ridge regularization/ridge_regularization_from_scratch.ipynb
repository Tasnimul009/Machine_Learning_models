```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 100,
                       n_features = 1,
                       n_targets = 1,
                       n_informative = 1,
                       noise = 20, 
                       random_state=0)
plt.scatter(X, y)
```

```python
class ridge_regularization :
    def __init__(self, alpha = 0.1):
        self.alpha = alpha
        self.m = None
        self.b = None
    def fit(self, X, y):
        num = 0
        den = 0
        num = np.sum((y-y.mean()) * (X.ravel()-X.mean()))
        den = np.sum((X.ravel()-X.mean())**2)
        self.m = num / (den + self.alpha)
        self.b = y.mean() - self.m * X.mean()
        print(self.m, self.b)
    def predict(self, X):
        y_pred = self.m*X+self.b
        return y_pred
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
reg.coef_
```

```python
model = ridge_regularization(alpha = 0.1)
model.fit(X, y)
plt.scatter(X, y, 
            edgecolor = "k",
            color = "yellow",
            s = 200)
plt.plot(X, 
         model.m*X+model.b, 
         color = "red",
         linewidth = 5)
```
