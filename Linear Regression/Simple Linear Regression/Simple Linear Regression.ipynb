```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("D:\Ml Models\Linear Regression\Dataset used\placement.csv")
df.head()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state = 0)
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
regressor.predict([X[0]])
```

```python
regressor.predict(X[0, 0].reshape(1, 1))
```

```python
plt.scatter(X_train, y_train, color = "cyan", edgecolor = "k", s= 200)
plt.plot(X_train, 
         regressor.predict(X_train),
         color = "red",
         linewidth = 2)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = "orange", edgecolor = "k", s=200)
plt.plot(X_train,
         regressor.predict(X_train),
         color = "red",
         linewidth = 2)
plt.show()
```
