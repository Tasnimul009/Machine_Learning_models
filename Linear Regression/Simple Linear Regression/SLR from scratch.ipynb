```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\Ml Models\Linear Regression\Dataset used\placement.csv")
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state=2)
X_train[:5]
y_train[:5]
```

```python
from SimpleLinearRegression import simpleLinearRegression
regressor = simpleLinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
plt.xlim(X[:, 0].min(), X[:, 0].max())
plt.ylim(y.min(), y.max())
plt.scatter(X, y,
            color = "yellow",
            edgecolor = "k",
            s = 200,
            label = "data points")
plt.plot(X_train,
         regressor.predict(X_train),
         color = "red",
         linewidth = 2,
         label = "regressor line")
plt.legend()
plt.show()
```
