```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("D:\Ml Models\Linear Regression\Dataset used\placement.csv")
df.head()
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel("CGPA")
plt.ylabel("package (int lakhs per annum)")
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state = 0)
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```
